# **Import Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# **Dataset Loading**

```python
df = pd.read_csv('https://raw.githubusercontent.com/itarasingh/AirBnb-Bookings-Analysis/refs/heads/main/Airbnb.csv')
```

# **Missing Values/Null Values**

```python
df.isnull().sum()
```

# **Clean Data**

```python
df.drop(['last_review', 'reviews_per_month'],axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

# **Univariate Analysis**

```python
plt.bar(df.index,height=df['number_of_reviews'])
plt.xlabel("Index of DataFrame")
plt.ylabel("Number of Reviews")
plt.title("Bar Chart of Number of Reviews")
plt.show()
```

```python
room_type_counts = df['room_type'].value_counts()
plt.pie(room_type_counts, labels = room_type_counts.index, autopct='%1.1f%%')
plt.show()
```

# **Bivariate analysis**

```python
correlation_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

# **Multivariate Analysis**

```python
plt.figure(figsize = (15, 15))
plt.subplot(221)
sns.scatterplot(x="latitude", y="longitude",hue="neighbourhood_group", data=df)
plt.subplot(222)
sns.scatterplot(x="latitude", y="longitude",hue="room_type", data=df)
plt.subplot(223)
sns.scatterplot(x="latitude", y="longitude",hue="price", data=df)
plt.subplot(224)
sns.scatterplot(x="latitude", y="longitude",hue="number_of_reviews", data=df)
plt.show()
```

# **The conclusion**
The majority of listings are for enitre home / apt and private rooms as shown by plot 2.
As expected, Manhattan prices are skewed towards the higher end (reateively sparse number of blue dots in the area) vs. the lower price points of other neighbourhood groups as shown in plot 3.
